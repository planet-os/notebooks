```python
from dh_py_access import package_api
import dh_py_access.lib.datahub as datahub
import xarray as xr
```

```python
API_key = open('APIKEY').read().strip()
server='api.planetos.com/'
version = 'v1'
```

```python
dataset_key = 'noaa_gfs_pgrb2_global_forecast_recompute_0.25degree'
variables = 'ugrd_m,vgrd_m'
area = 'Australia_New_zealand2'
latitude_north = -29.5;  latitude_south = -50
longitude_west = 140; longitude_east = 180
```

```python
dh=datahub.datahub(server,version,API_key)
package = package_api.package_api(dh,dataset_key,variables,longitude_west,longitude_east,latitude_south,latitude_north,area_name=area,z=0)
package.make_package()
package.download_package()
```

```python
dd = xr.open_dataset(package.local_file_name)
```
